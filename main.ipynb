```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_table('2004-2021.tsv')

display(df.head(5))
display(df.info())
```

```python


# Colunas de distribui√ß√£o que vieram como object mas s√£o n√∫meros
cols_dist = [
    'PRE√áO M√âDIO DISTRIBUI√á√ÉO',
    'DESVIO PADR√ÉO DISTRIBUI√á√ÉO',
    'PRE√áO M√çNIMO DISTRIBUI√á√ÉO',
    'PRE√áO M√ÅXIMO DISTRIBUI√á√ÉO',
    'COEF DE VARIA√á√ÉO DISTRIBUI√á√ÉO',
    'MARGEM M√âDIA REVENDA'
]

for col in cols_dist:
    # Substituir '-' por NaN e depois converter v√≠rgula para ponto
    df[col] = df[col].replace('-', None)  # ou np.nan
    df[col] = df[col].str.replace(',', '.', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.drop(['DATA INICIAL', 'DATA FINAL'], axis=1)


df.info()
display(df.head(5))
```

```python

#Grafico de porcentagens de produtos mais destribuidos e de regioes

# ======== C√ÅLCULOS ========
# Porcentagem de registros
porcentagem_produto = df['PRODUTO'].value_counts(normalize=True) * 100
porcentagem_regiao = df['REGI√ÉO'].value_counts(normalize=True) * 100
porcentagem_regiao = porcentagem_regiao.sort_values()

# Tabela cruzada (pivot table)
tabela_preco = df.pivot_table(
    values='PRE√áO M√âDIO REVENDA',
    index='PRODUTO',
    columns='REGI√ÉO',
    aggfunc='mean'
)

# Pre√ßos m√©dios
preco_produto = df.groupby('PRODUTO')['PRE√áO M√âDIO REVENDA'].mean().sort_values(ascending=False)
preco_regiao = df.groupby('REGI√ÉO')['PRE√áO M√âDIO REVENDA'].mean().sort_values(ascending=False)

# ======== ESTILO ========
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
plt.suptitle('An√°lises Percentuais e de Pre√ßo M√©dio por Produto e Regi√£o',
             fontsize=16, fontweight='bold', y=0.98)

# Paletas de cores
cores_produto = plt.cm.Blues(np.linspace(1, 0.3, len(porcentagem_produto)))
cores_regiao = plt.cm.Oranges(np.linspace(1, 0.3, len(porcentagem_regiao)))
cores_preco_produto = plt.cm.Purples(np.linspace(0.7, 0.2, len(preco_produto)))
cores_preco_regiao = plt.cm.Greens(np.linspace(0.7, 0.2, len(preco_regiao)))


# ======== GR√ÅFICO 1: Distribui√ß√£o por Produto ========
porcentagem_produto.plot(kind='bar', color=cores_produto, edgecolor='black', ax=axes[0, 0])
axes[0, 0].set_title('Distribui√ß√£o de Registros por Produto', fontsize=13, fontweight='bold', pad=10)
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('Quantidade (%)', fontsize=12)
axes[0, 0].tick_params(axis='x', rotation=45)

for p in axes[0, 0].patches:
    axes[0, 0].text(p.get_x() + p.get_width()/2, p.get_height() + 0.5,
                    f'{p.get_height():.1f}%', ha='center', va='bottom', fontsize=9.5, fontweight='bold')

axes[0, 0].spines[['right', 'top']].set_visible(False)
axes[0, 0].grid(alpha=0.3, axis='y')


# ======== GR√ÅFICO 2: Distribui√ß√£o por Regi√£o ========
porcentagem_regiao.plot(kind='barh', color=cores_regiao, edgecolor='black', ax=axes[0, 1])
axes[0, 1].set_title('Distribui√ß√£o de Registros por Regi√£o', fontsize=13, fontweight='bold', pad=10)
axes[0, 1].set_xlabel('Quantidade (%)', fontsize=12)
axes[0, 1].set_ylabel('')

for p in axes[0, 1].patches:
    width = p.get_width()
    y = p.get_y() + p.get_height()/2
    axes[0, 1].text(width + 0.3, y, f'{width:.1f}%', ha='left', va='center',
                    fontsize=9.5, fontweight='bold', color='#333333')

axes[0, 1].spines[['right', 'top']].set_visible(False)
axes[0, 1].grid(alpha=0.3, axis='x')


# ======== GR√ÅFICO 3: Pre√ßo M√©dio por Produto ========
preco_produto.plot(kind='bar', color=cores_preco_produto, edgecolor='black', ax=axes[1, 0])
axes[1, 0].set_title('Pre√ßo M√©dio de Revenda por Produto (R$)', fontsize=13, fontweight='bold', pad=10)
axes[1, 0].set_xlabel('Produto', fontsize=12)
axes[1, 0].set_ylabel('Pre√ßo M√©dio (R$)', fontsize=12)
axes[1, 0].tick_params(axis='x', rotation=45)

for p in axes[1, 0].patches:
    axes[1, 0].text(p.get_x() + p.get_width()/2, p.get_height() + 0.02,
                    f'R${p.get_height():.2f}', ha='center', va='bottom', fontsize=9, fontweight='bold')



# ======== GR√ÅFICO 4: Pre√ßo M√©dio por Regi√£o ========
preco_regiao.plot(kind='barh', color=cores_preco_regiao, edgecolor='black', ax=axes[1, 1])
axes[1, 1].set_title('Pre√ßo M√©dio de Revenda por Regi√£o (R$)', fontsize=13, fontweight='bold', pad=10)
axes[1, 1].set_xlabel('Pre√ßo M√©dio (R$)', fontsize=12)
axes[1, 1].set_ylabel('')

for p in axes[1, 1].patches:
    width = p.get_width()
    y = p.get_y() + p.get_height()/2
    axes[1, 1].text(width + 0.03, y, f'R${width:.2f}', ha='left', va='center',
                    fontsize=9, fontweight='bold', color='#333333')

axes[1, 1].spines[['right', 'top']].set_visible(False)
axes[1, 1].grid(alpha=0.3, axis='x')

# ======== AJUSTES FINAIS ========
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()




plt.figure(figsize=(10, 6))
sns.heatmap(tabela_preco, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Pre√ßo M√©dio: Produto x Regi√£o')
plt.show()
```

```python
import geopandas as gpd

# Carregar mapa
url = 'https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2021/Brasil/BR/BR_UF_2021.zip'
brasil = gpd.read_file(url)

# Filtrar dados sem GLP
df_sem_glp = df[df['PRODUTO'] != 'GLP']
preco_por_estado = df_sem_glp.groupby('ESTADO')['PRE√áO M√âDIO REVENDA'].mean().sort_values(ascending=False).reset_index()

# Dicion√°rio de siglas
siglas = {
    'ACRE': 'AC', 'RORAIMA': 'RR', 'RONDONIA': 'RO', 'ROND√îNIA': 'RO',
    'PARA': 'PA', 'PAR√Å': 'PA', 'AMAPA': 'AP', 'AMAP√Å': 'AP',
    'TOCANTINS': 'TO', 'DISTRITO FEDERAL': 'DF', 'PIAUI': 'PI', 'PIAU√ç': 'PI',
    'MARANHAO': 'MA', 'MARANH√ÉO': 'MA', 'AMAZONAS': 'AM', 'RIO GRANDE DO SUL': 'RS',
    'MATO GROSSO': 'MT', 'GOIAS': 'GO', 'GOI√ÅS': 'GO', 'CEARA': 'CE', 'CEAR√Å': 'CE',
    'ALAGOAS': 'AL', 'RIO GRANDE DO NORTE': 'RN', 'MATO GROSSO DO SUL': 'MS',
    'SERGIPE': 'SE', 'RIO DE JANEIRO': 'RJ', 'ESPIRITO SANTO': 'ES', 'ESP√çRITO SANTO': 'ES',
    'BAHIA': 'BA', 'SANTA CATARINA': 'SC', 'MINAS GERAIS': 'MG',
    'PARAIBA': 'PB', 'PARA√çBA': 'PB', 'PERNAMBUCO': 'PE',
    'PARANA': 'PR', 'PARAN√Å': 'PR', 'SAO PAULO': 'SP', 'S√ÉO PAULO': 'SP'
}

# Adicionar siglas
preco_por_estado['SIGLA'] = preco_por_estado['ESTADO'].map(siglas)

# Merge com shapefile
brasil = brasil.merge(preco_por_estado, left_on='SIGLA', right_on='SIGLA', how='left')

# ======== CRIAR SUBPLOTS ========
fig = plt.figure(figsize=(20, 12))
gs = fig.add_gridspec(1, 2, width_ratios=[2.5, 1], wspace=0.15)

# ======== SUBPLOT 1: MAPA ========
ax1 = fig.add_subplot(gs[0])
brasil.plot(column='PRE√áO M√âDIO REVENDA', ax=ax1, 
            legend=False,
            cmap='RdYlGn_r', edgecolor='black', linewidth=0.8,
            missing_kwds={'color': 'lightgrey'})

# Adicionar siglas diretamente
for idx, row in brasil.iterrows():
    centroid = row['geometry'].centroid
    ax1.text(centroid.x, centroid.y, row['SIGLA'], 
             fontsize=8, ha='center', va='center', 
             fontweight='bold', color='white',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='black', alpha=0.6))


ax1.set_title('Mapa de Pre√ßos de Combust√≠vel no Brasil', 
              fontsize=16, fontweight='bold', pad=15)
ax1.axis('off')

# ======== SUBPLOT 2: HEATMAP ========
ax2 = fig.add_subplot(gs[1])

# Preparar dados para heatmap (ordenado por pre√ßo)
preco_ordenado = preco_por_estado.set_index('SIGLA')['PRE√áO M√âDIO REVENDA'].sort_values(ascending=False)
matriz = preco_ordenado.values.reshape(-1, 1)

# Criar heatmap
sns.heatmap(matriz, annot=True, fmt='.2f', cmap='RdYlGn_r',
            yticklabels=preco_ordenado.index,
            xticklabels=['Pre√ßo M√©dio'],
            cbar_kws={'label': 'Pre√ßo (R$/L)'}, 
            ax=ax2, linewidths=0.5, linecolor='gray')

ax2.set_title('Pre√ßo M√©dio por Estado\n(Sem GLP)', 
              fontsize=14, fontweight='bold', pad=15)
ax2.set_ylabel('')

# T√≠tulo geral
plt.suptitle('An√°lise Geogr√°fica de Pre√ßos de Combust√≠vel no Brasil (2004-2021)', 
             fontsize=20, fontweight='bold', y=0.98)

plt.tight_layout()
plt.savefig('mapa_heatmap_combustivel.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
gasolina = df[df['PRODUTO'] == 'GASOLINA COMUM']['PRE√áO M√âDIO REVENDA'].mean()
etanol = df[df['PRODUTO'] == 'ETANOL HIDRATADO']['PRE√áO M√âDIO REVENDA'].mean()
razao = etanol/ gasolina

gasolina_por_estado = df[df['PRODUTO'] == 'GASOLINA COMUM'].groupby('ESTADO')['PRE√áO M√âDIO REVENDA'].mean()
etanol_por_estado = df[df['PRODUTO'] == 'ETANOL HIDRATADO'].groupby('ESTADO')['PRE√áO M√âDIO REVENDA'].mean()

razao_estado = (etanol_por_estado / gasolina_por_estado).reset_index()
razao_estado.columns = ['ESTADO', 'RAZAO']  # Renomear coluna

#-------------------------------------------------------------------------------------------------------------------
import matplotlib.patches as mpatches

plt.figure(figsize=(10, 5))  # Aumentar largura para caber legenda

razao_ordenado = razao_estado.sort_values('RAZAO')
cores = ['green' if x < 0.70 else 'blue' for x in razao_ordenado['RAZAO']]

plt.bar(razao_ordenado['ESTADO'], razao_ordenado['RAZAO'], 
        color=cores, edgecolor='black', alpha=0.7)

plt.axhline(y=0.70, color='darkred', linestyle='--', linewidth=2)

plt.xlabel('Estado', fontsize=12, fontweight='bold')
plt.ylabel('Raz√£o Etanol/Gasolina', fontsize=12, fontweight='bold')
plt.title('Viabilidade do (Etanol X Gasolina) por Estado (2004-2021)', fontsize=14, fontweight='bold')
plt.xticks(rotation=90, fontsize=9)
plt.grid(alpha=0.5, axis='y', linestyle='--')

#  Legenda do lado direito, fora
verde_patch = mpatches.Patch(color='green', label=' Etanol vale a pena', alpha=0.7)
vermelho_patch = mpatches.Patch(color='blue', label=' Gasolina vale a pena', alpha=0.7)

plt.legend(handles=[verde_patch, vermelho_patch], 
           fontsize=11, 
           loc='center left',
           bbox_to_anchor=(1.01, 0.5),
           frameon=True,
           fancybox=True,
           shadow=True)

plt.tight_layout()
plt.show()
```

### 4.2 Variabilidade de Pre√ßos
**Perguntas:**
- Onde os pre√ßos s√£o mais inst√°veis?
- Qual produto tem maior volatilidade?
- O coeficiente de varia√ß√£o est√° relacionado com a regi√£o?

**Gr√°ficos:**
- Boxplot do coeficiente de varia√ß√£o
- Barras dos estados com maior/menor varia√ß√£o
- Correla√ß√£o entre vari√°veis


```python


# Qual produto tem maior volatilidade?
volatilidade_produto = df.groupby('PRODUTO')['COEF DE VARIA√á√ÉO REVENDA'].mean()



# O coeficiente de varia√ß√£o est√° relacionado com a regi√£o?

df_sem_glp = df[df['PRODUTO'] != 'GLP']
volatilidade_regiao = (
    df_sem_glp
    .groupby('REGI√ÉO')['COEF DE VARIA√á√ÉO REVENDA']
    .mean()
    .sort_values(ascending=True)

)



preco_regiao = (
    df_sem_glp
    .groupby('REGI√ÉO')['PRE√áO M√âDIO REVENDA']
    .mean()
    .sort_values(ascending=True)    
    )

preco_regiao = preco_regiao.reindex(volatilidade_regiao.index)




ffig, ax1 = plt.subplots(figsize=(10,6))

# Eixo 1 - Coeficiente de varia√ß√£o
ax1.bar(volatilidade_regiao.index, volatilidade_regiao.values,
        color='coral', alpha=0.7, label='Coef. de Varia√ß√£o')
ax1.set_ylabel('Coeficiente de Varia√ß√£o', color='coral', fontsize=11)
ax1.tick_params(axis='y', labelcolor='coral')
ax1.set_xticklabels(volatilidade_regiao.index, rotation=45)

# Eixo 2 - Pre√ßo m√©dio (segundo eixo Y)
ax2 = ax1.twinx()
ax2.plot(preco_regiao.index, preco_regiao.values, 
         color='steelblue', marker='o', linewidth=2, label='Pre√ßo M√©dio')
ax2.set_ylabel('Pre√ßo M√©dio (R$/L)', color='steelblue', fontsize=11)
ax2.tick_params(axis='y', labelcolor='steelblue')

# T√≠tulo
plt.title('Volatilidade e Pre√ßo M√©dio por Regi√£o', fontsize=14, fontweight='bold')

# Legendas combinadas
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()


```

Regi√µes como o Norte apresentam pre√ßos m√©dios mais altos e maior volatilidade, enquanto o Sul se destaca pela estabilidade e pre√ßos mais baixos.
Isso indica que diferen√ßas regionais em log√≠stica, concorr√™ncia e tributos influenciam tanto o valor quanto a varia√ß√£o dos pre√ßos de revenda.
